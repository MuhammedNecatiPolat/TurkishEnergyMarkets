```python
import pandas as pd

data = pd.read_csv('./data/data2019.csv')
```

```python
data.head()
```

```python
# Rename columns accroding to the Prophet requirements
data = data.rename(columns={'date': 'ds', 'netImbalanceVolume': 'y'})
# Remove timezone from the date
data['ds'] = data['ds'].apply(lambda x: x[:-6])
data['ds'] = pd.to_datetime(data['ds'])
data.head()
```

```python
from prophet import Prophet
m = Prophet()
```

```python
m.fit(data)
```

```python
future = m.make_future_dataframe(periods=24, freq='H')
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)
```
